# **Visualization**

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['generated_power_kw'], bins=30, kde=True)
plt.xlabel('generated_power_kw')
plt.ylabel('Frequency')
plt.title('Distribution of generated_power_kw')
plt.show()
```
